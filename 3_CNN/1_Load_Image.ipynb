# 图像加载

## 灰色图片

```python
import matplotlib.pyplot as plt
import numpy as np

img1 = np.zeros([200, 300, 3]) # 全0，黑色图像
plt.imshow(img1)
plt.show()

img2 = np.full([200, 300, 3], 255) # 白色图像
plt.imshow(img2)
plt.show()

img3 = np.full([200, 300, 3], 128) # 灰色图像
plt.imshow(img3)
plt.show()
```
